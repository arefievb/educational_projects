## Расчёты и добавление результатов в таблицу

Cчитаю и добавляю в таблицу цену квадратного метра (price_per_meter).  
Для этого разделю значение из столбца last_price (цена на момент снятия с публикации) на столбец  
total_area (площадь квартиры в квадратных метрах (м²))


```python
#Создаю новый стобец price_per_meter, заполняю значениями
#Сразу перевожу в тип int
data['price_per_meter'] = (data['last_price'] / data['total_area']).astype(int)
#Проверяю созданный столбец
data['price_per_meter'].head()
```

Высчитываю день недели, месяц и год публикации объявления.  
С помощью данных из столбца first_day_exposition (дата публикации).

```python
#добавляю столбец exposition_weekday, в нем будут хранится номера дней недели от 0 до 6
data['exposition_weekday'] = data['first_day_exposition'].dt.weekday

#Меняю численные значения дней недели на более понятные для заказчика
data['exposition_weekday'] = data['exposition_weekday'].replace({0 : 'Понедельник',
                                                                 1 : 'Вторник',
                                                                 2 : 'Среда',
                                                                 3 : 'Четверг',
                                                                 4 : 'Пятница',
                                                                 5 : 'Суббота',
                                                                 6 : 'Воскресенье',
                                                                })
data['exposition_weekday'].astype('category')

#добавляю столбец exposition_month, в нем будут хранится номера месяцев
data['exposition_month'] = data['first_day_exposition'].dt.month

#добавляю столбец exposition_year, в нем будут хранится года
data['exposition_year'] = data['first_day_exposition'].dt.year

#проверяю добавленный столбцы
data[['exposition_weekday', 'exposition_month', 'exposition_year','first_day_exposition']].head(5)
```

Категоризирую этаж квартиры по 3 группам: Первый, последний, другой.  
Для этого мы воспользуемся данными из столбцов floor (этаж) и floors_total (всего этажей в доме)

```python
#Создаю функцию, принимающую как аргумент ряд
def floor_kind(row):
    floor = row['floor']
    floors_total = row['floors_total']
#Внутри функции создаю переменные из ряда    
    if floor == 1:
        return 'Первый'
#Если переменная floor равна 1, то функция возвращает строку 'Первый'     
    elif floor == floors_total:
        return 'Последний'
#Если переменная floor равна переменной floors_total, то функция возвращает строку 'Последний' 
    else:
        return 'Другой'
#В остальных случаях, когда этаж не первый и не последний, функция возвращает строку 'Другой'

#Проверим работу функции
#Создаю два списка. В одном — 3 cтроки значений, в другом — названия столбцов датафрейма.
row_floor = [2,3]
row_columns = ['floor', 'floors_total']
#Формирую строку
row = pd.Series(data = row_floor, index = row_columns)
#Передаю строку в функцию в качетсве аргумента
floor_kind(row)
```

Функция работает корректно, применим ее к датафрейму c помощью метода apply.  
Вместе с функцией передаем аргумент axis = 1, чтобы на вход функции передавались строки а не столбцы.

```python
#Применяю функцию
data['floor_kind'] = data.apply(floor_kind, axis = 1)

#Проверяю созданный столбец
data['floor_kind'].value_counts()
```

Создаем новые столбцы со следующими параметрами:  
- отношение жилой к общей площади (living_ratio)
- отношение площади кухни к общей площади (kitchen_ratio)

```python
#сразу округлим до 3 знаков после запятой
data['living_ratio'] = (data['living_area']/ data['total_area']).round(3)
data['kitchen_ratio'] = (data['kitchen_area']/data['total_area']).round(3)
data[['kitchen_ratio', 'living_ratio']].head()
```

### Вывод
- Добавлена метрика "Цена квадратного метра" в столбце "price_per_meter", значения переведены в тип int, так как для данного параметра знаки после запятой имеют крайне малое значение.
- Определены день недели, месяц и год публикации объявления (exposition_weekday, exposition_month, exposition_year соответственно). Дни недели представлены в стандартном виде для удобства.
- Произведена категоризация по значению "этаж квартиры"; варианты — первый, последний, другой. Для категорий создан отдельный столбец - 'floor_kind'
- Определены соотношение жилой площади и площади кухни к общей площади ('kitchen_ratio', 'living_ratio'). Значения округлены до 3ех знаков после запятой для удобства восприятия.
